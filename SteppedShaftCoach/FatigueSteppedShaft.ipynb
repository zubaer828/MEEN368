```python
from numpy import *
import matplotlib.pyplot as plt

# Geometry and material information
S_ut = 120; S_yt = 66;
L = 20; a = 15; b = 5; c = 10
d = 2.0;

# Step-1: Determine sigma_a and sigma_m
F = 8;
R1 = F*b/L;
M = R1*c;
sqrta = 0.246 - 3.08e-3*S_ut + 1.51e-5*(S_ut**2) - 2.67e-8*(S_ut**3)
r = 0.1*d;
Kt = 1.65;
Kf = 1 + (Kt -1)/(1+sqrta/sqrt(r))

sigma_max = Kf*32*M/(pi*d**3)
sigma_min = -Kf*32*M/(pi*d**3)
sigma_a = (sigma_max - sigma_min)/2
sigma_m = (sigma_max + sigma_min)/2
print("M =",M)
print("Kt =",Kt, "sqrt(a) =", sqrta)
print("Kf =", Kf)
print("Sigma_a (from applied loading) = ", sigma_a)


# Step-2: Determine Se
ka = 2.0*S_ut**(-0.217)
kb = 0.91*d**(-0.157)
kc = 1; kd = 1; ke = 1
Se_prime = 0.5*S_ut
Se = ka * kb * kc * kd * ke * Se_prime
print("Se = ",Se)
# Step-3: Determine nf
if sigma_m <= 0:
    n_f = Se/sigma_a
elif sigma_m > 0:
    n_f = (sigma_a/Se + sigma_m/S_ut)**(-1)
print("FOS: n_f =", n_f)


# Step-4: Life-related calculation
N = 950*60*10;
f = 1.06-2.8e-3*S_ut + 6.9e-6*S_ut**2
a = (f*S_ut)**2/Se
b = -(1/3)*log10(f*S_ut/Se)
sigma_aN = a*N**b;
print("a =", a, "b =", b, "f=",f)
print("Sigma_a (for expected N cycles) = ",sigma_aN)
print("FOS: n_d =",sigma_aN/sigma_a)

```
